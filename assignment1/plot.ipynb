```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""TEST 1: L = 8, M = 4"""

with open('x_L8_M4.txt', 'r') as f:
    x = f.read().split("\n")


with open('ecgsig.txt', 'r') as f:
    ecgsig = f.read().split("\n")


plt.rcParams['figure.figsize'] = [10,5]
plt.title("L = 8, M = 4")
plt.plot(np.array(ecgsig).astype(float), 'b')
plt.plot(np.array(x).astype(float),  'r')
plt.legend(["original", "restored"])

print("Time for compression incl. H = 1765 [m/s]")
print("Time for compression W/O. H = 1203 [m/s]")
```

```python
"""TEST 2: L = 16, M = 4"""

with open('x_L16_M4.txt', 'r') as f:
    x = f.read().split("\n")


with open('ecgsig.txt', 'r') as f:
    ecgsig = f.read().split("\n")


plt.rcParams['figure.figsize'] = [10,5]
plt.title("L = 16, M = 4")
plt.plot(np.array(ecgsig).astype(float), 'b')
plt.plot(np.array(x).astype(float),  'r')
plt.legend(["original", "restored"])

print("Time for compression incl. H = 6023 [m/s]")
print("Time for compression W/O. H = 3609 [m/s]")
```

As expected it takes longer to generate H and compress as we increase L.
