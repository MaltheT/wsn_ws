```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


plt.rcParams['figure.figsize'] = [10, 5]

filename_const = './results/constant_jammer.txt'
filename_random = './results/random_jammer.txt'
filename_test = './results/test.txt'

#CONST:
#-66 	 -97 	 -88
#changed to channel 17 because of RSSI

#RANDOM
#-54;-95;-86
#changed to channel 17 because of RSSI
```

```python
data_const = pd.read_csv(filename_const, sep=";")
data_random = pd.read_csv(filename_random, sep=";")
data_test = pd.read_csv(filename_test, sep=";")


print(data_const)
```
